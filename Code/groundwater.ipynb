```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# coverting data from the excel sheet 
data = {
    'Block': ['Minjur', 'Cholavaram', 'Puzhal', 'Villivakkam', 'Ponnamalle', 'Sriperumbudur', 'Kundrathur', 'St. Thomas Mount', 'Kattankulathur'],
    'Total Area (km2)': [478.30, 193.69, 60.41, 175.78, 156.13, 365.69, 270.38, 236.51, 361.76],
    'Total Annual Groundwater Recharge (Mm3)': [123.72, 98.40, 34.87, 60.65, 72.01, 134.03, 87.66, 41.61, 83.40],
    'Net Annual Groundwater Availability (Mm3)': [111.35, 88.56, 31.38, 54.59, 64.81, 120.62, 78.90, 37.45, 75.06],
    'Existing Gross Groundwater Draft for All Uses (Mm3)': [147.31, 68.43, 16.01, 28.72, 57.39, 23.74, 45.26, 23.85, 45.49],
    'Stage of Groundwater Development (%)': [132, 77, 51, 53, 89, 20, 57, 64, 61]
}

# Convert data to numpy arrays
block = np.array(data['Block'])
total_area = np.array(data['Total Area (km2)'])
total_recharge = np.array(data['Total Annual Groundwater Recharge (Mm3)'])
net_availability = np.array(data['Net Annual Groundwater Availability (Mm3)'])
existing_draft = np.array(data['Existing Gross Groundwater Draft for All Uses (Mm3)'])
stage_of_dev = np.array(data['Stage of Groundwater Development (%)'])

# Create bar chart for total area
fig, ax = plt.subplots()
ax.bar(block, total_area)
ax.set_title('Total Area by Block')
ax.set_xlabel('Block')
ax.set_ylabel('Total Area (km2)')
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.25)
plt.show()

# Create stacked bar chart for groundwater data
fig, ax = plt.subplots()
ax.bar(block, total_recharge, label='Total Annual Groundwater Recharge (Mm3)')
ax.bar(block, net_availability, bottom=total_recharge, label='Net Annual Groundwater Availability (Mm3)')
ax.bar(block, existing_draft, bottom=total_recharge+net_availability, label='Existing Gross Groundwater Draft for All Uses (Mm3)')
ax.set_title('Groundwater Data by Block')
ax.set_xlabel('Block')
ax.set_ylabel('Groundwater (Mm3)')
ax.legend()
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.25)
plt.show()

```
